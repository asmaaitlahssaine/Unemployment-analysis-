```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
data = pd.read_csv("C:/Users/Lenovo/Unemployment Analysis with Python/Unemployment_Rate_upto_11_2020.csv")
```

```python
print(data.head())
```

```python
print(data.isnull().sum())
```

```python
# Filter only numeric columns for correlation
numerical_data = data.select_dtypes(include=['float64', 'int64'])
from matplotlib.colors import LinearSegmentedColormap
custom_cmap = LinearSegmentedColormap.from_list("Indigo_Magenta", ["Indigo", "Magenta"])

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(numerical_data.corr(), annot=True, cmap=custom_cmap)
plt.show()
```

```python
data.columns=["States", "Date", "Frequency", "Estimated Unemployment Rate", "Estimated Employed", "Estimated Labour Participation Rate", "Region", "Longitude", "Latitude"]
plt.title("India Unemployment ")
sns.histplot(x="Estimated Employed", hue="Region", data=data)
plt.show()
```

```python
plt.figure(figsize=(12,10))
plt.title("Indian Unemployement")
sns.histplot(x="Estimated Unemployment Rate", hue="Region", data=data)
plt.show()
```

```python
# Set the style for the plot
sns.set(style="whitegrid")

# Create a bar plot with the updated approach
plt.figure(figsize=(10, 6))
sns.barplot(
    x="Region",
    y="Estimated Unemployment Rate",
    data=data,
    estimator=sum,
    hue="Region",  # Assigning the x variable to hue
    dodge=False,   # Disable dodging to ensure single bars
    palette="viridis",
    legend=False   # Disable the legend
)

# Set the title and labels
plt.title("Unemployment Rate by Region in India")
plt.xlabel("Region")
plt.ylabel("Total Estimated Unemployment Rate")
plt.xticks(rotation=45)  # Rotate x labels for better readability

# Show the plot
plt.show()
```
